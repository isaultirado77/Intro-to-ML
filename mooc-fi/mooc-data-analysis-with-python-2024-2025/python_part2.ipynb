## Regular expresions

```python
import re
```

```python
def is_match(m): 
    if m: 
        print("Match")
        print("group: ", m.group())
        print("start: ", m.start(), "end: ", m.end())
    else: 
        print("Not Match")
```

```python
w = "Banana"
m = re.match(pattern=r"[Aa]pple|[Bb]anana", string=w)
is_match(m)
```

```python
validname = r"[A-Za-z_][A-Za-z_0-9]*\Z"
w1 = "A123_"
w2 = "_hidden"
w3 = "2abc"
m1 = re.match(validname, w1)
is_match(m1)
m2 = re.match(validname, w2)
is_match(m2)
m3 = re.match(validname, w3)
is_match(m3)
```

## File processing & file objects


```python
f = open("data/text.txt", "r")
for i in range(5): 
    print(f.readline())
```

```python
# Using a context manager
with open("data/text.txt", "r") as f: 
    for i in range(5): 
        print(f.readline())
```

```python
with open("data/text.txt", "r") as f:
    for line in f: 
        print(line)
```

```python
max_len = -1
with open("data/text.txt", "r") as f:
    for line in f: 
        if len(line) > max_len:
            max_len = len(line)
print("Lingest line: ", max_len)
```
